```python
import torch.nn as nn
import timm

class DualStreamModel(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()

        # CNN backbones
        self.cnn_full = timm.create_model('resnet50', pretrained=True, num_classes=0)
        self.cnn_roi  = timm.create_model('resnet18', pretrained=True, num_classes=0)

        # Fix dimensions
        self.full_fc = nn.Linear(2048, 512)
        self.roi_fc  = nn.Linear(512, 512)

        # Transformer
        encoder_layer = nn.TransformerEncoderLayer(d_model=512, nhead=8)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)

        # Attention fusion
        self.attention = nn.Sequential(
            nn.Linear(1024, 128),
            nn.ReLU(),
            nn.Linear(128, 2),
            nn.Softmax(dim=1)
        )

        # Heads
        self.classifier = nn.Linear(512, num_classes)
        self.severity   = nn.Linear(512, 1)
        self.count_head = nn.Linear(512, 1)

    def forward(self, img, roi):
        f_full = self.cnn_full(img)
        f_roi  = self.cnn_roi(roi)

        f_full = self.full_fc(f_full)
        f_roi  = self.roi_fc(f_roi)

        # Transformer on full features
        f_full = self.transformer(f_full.unsqueeze(0)).squeeze(0)

        # Fusion
        combined = torch.cat([f_full, f_roi], dim=1)
        weights = self.attention(combined)

        fused = weights[:, 0:1] * f_full + weights[:, 1:2] * f_roi

        cls = self.classifier(fused)
        sev = self.severity(fused)
        cnt = self.count_head(fused)

        return cls, sev, cnt
```

```python
def evaluate(loader):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for img, label in loader:
            img, label = img.to(device), label.to(device)

            # ROI from DeepLab
            roi, _ = get_roi_from_deeplab(img)

            cls, _, _ = model(img, roi)

            pred = cls.argmax(dim=1)

            correct += (pred == label).sum().item()
            total += label.size(0)

    return correct / total
```

```python
import cv2
import numpy as np
import torch

def count_lesions_from_mask(mask_batch):
    """
    mask_batch: (B,1,H,W) tensor
    returns: (B,1) tensor with lesion counts
    """
    counts = []
    masks_np = mask_batch.detach().cpu().numpy()

    for i in range(masks_np.shape[0]):
        m = masks_np[i, 0]
        m = (m > 0).astype(np.uint8) * 255

        # clean small noise
        kernel = np.ones((3,3), np.uint8)
        m = cv2.morphologyEx(m, cv2.MORPH_OPEN, kernel)

        num_labels, _ = cv2.connectedComponents(m)
        counts.append(num_labels - 1)  # remove background

    return torch.tensor(counts, dtype=torch.float32).unsqueeze(1)
```

```python
def dice_score(pred, target):
    pred = (pred > 0).float()
    target = (target > 0).float()

    intersection = (pred * target).sum()
    union = pred.sum() + target.sum()

    return (2.0 * intersection) / (union + 1e-8)
```

```python
def compute_severity(mask):
    pixel_count = mask.sum(dim=[1,2,3])
    severity = torch.zeros_like(pixel_count)

    severity[pixel_count < 5000] = 1
    severity[(pixel_count >= 5000) & (pixel_count < 15000)] = 2
    severity[pixel_count >= 15000] = 3

    return severity.float().unsqueeze(1)


def count_lesions_from_mask(mask):
    counts = []
    masks_np = mask.detach().cpu().numpy()

    for i in range(masks_np.shape[0]):
        m = (masks_np[i,0] > 0).astype(np.uint8) * 255
        num_labels, _ = cv2.connectedComponents(m)
        counts.append(num_labels - 1)

    return torch.tensor(counts, dtype=torch.float32).unsqueeze(1)
```

```python
class DualStreamModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.cnn_full = timm.create_model('resnet50', pretrained=True, num_classes=0)
        self.cnn_roi  = timm.create_model('resnet18', pretrained=True, num_classes=0)

        self.full_fc = nn.Linear(2048, 512)
        self.roi_fc  = nn.Linear(512, 512)

        encoder_layer = nn.TransformerEncoderLayer(d_model=512, nhead=8, batch_first=True)
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=2)

        self.attention = nn.Sequential(
            nn.Linear(1024, 128),
            nn.ReLU(),
            nn.Linear(128, 2),
            nn.Softmax(dim=1)
        )

        self.classifier = nn.Linear(512, 2)
        self.severity   = nn.Linear(512, 1)
        self.count_head = nn.Linear(512, 1)

    def forward(self, img, roi):
        f_full = self.cnn_full(img)
        f_roi  = self.cnn_roi(roi)

        f_full = self.full_fc(f_full)
        f_roi  = self.roi_fc(f_roi)

        f_full = self.transformer(f_full.unsqueeze(1)).squeeze(1)

        combined = torch.cat([f_full, f_roi], dim=1)
        weights = self.attention(combined)

        fused = weights[:,0:1]*f_full + weights[:,1:2]*f_roi

        cls = self.classifier(fused)
        sev = self.severity(fused)
        cnt = self.count_head(fused)

        return cls, sev, cnt
```

```python
def train_epoch(loader):
    model.train()
    total_loss = 0

    for img, label in loader:
        img, label = img.to(device), label.to(device)

        roi, mask = get_roi_from_deeplab(img)

        optimizer.zero_grad()

        with autocast("cuda"):
            cls, sev, cnt = model(img, roi)

            sev_target = compute_severity(mask).to(device)
            cnt_target = count_lesions_from_mask(mask).to(device)

            loss = (
                criterion_cls(cls, label)
                + 0.7*criterion_reg(sev, sev_target)
                + 0.3*criterion_reg(cnt, cnt_target)
            )

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()

    return total_loss / len(loader)
```

```python
import torch
import torch.nn as nn
import timm
import torch.nn.functional as F
from torch.amp import autocast, GradScaler

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ✅ MODEL
class SimpleDualModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.cnn_full = timm.create_model('mobilenetv3_small_100', pretrained=True, num_classes=0)
        self.cnn_roi  = timm.create_model('mobilenetv3_small_100', pretrained=True, num_classes=0)

        self.fc = nn.Linear(1024, 256)

        self.classifier = nn.Linear(256, 2)
        self.severity   = nn.Linear(256, 1)
        self.count_head = nn.Linear(256, 1)

    def forward(self, img, roi):
        f_full = self.cnn_full(img)
        f_roi  = self.cnn_roi(roi)

        fused = torch.cat([f_full, f_roi], dim=1)
        fused = self.fc(fused)

        cls = self.classifier(fused)
        sev = self.severity(fused)
        cnt = self.count_head(fused)

        return cls, sev, cnt


model = SimpleDualModel().to(device)

# ✅ LOSSES
weights = torch.tensor([3.0, 1.0]).to(device)
criterion_cls = nn.CrossEntropyLoss(weight=weights)
criterion_reg = nn.MSELoss()

optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
scaler = GradScaler("cuda")


# ✅ TRAIN FUNCTION
def train_epoch(loader):
    model.train()
    total_loss = 0

    for img, label in loader:
        img, label = img.to(device), label.to(device)

        roi, mask = get_roi_from_deeplab(img)

        optimizer.zero_grad()

        with autocast("cuda"):
            cls, sev, cnt = model(img, roi)

            sev_target = compute_severity(mask).to(device)
            cnt_target = count_lesions_from_mask(mask).to(device)

            loss = (
                criterion_cls(cls, label)
                + 0.7*criterion_reg(sev, sev_target)
                + 0.3*criterion_reg(cnt, cnt_target)
            )

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()

    return total_loss / len(loader)


# ✅ EVALUATE FUNCTION
def evaluate(loader):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for img, label in loader:
            img, label = img.to(device), label.to(device)

            roi, _ = get_roi_from_deeplab(img)
            cls, _, _ = model(img, roi)

            pred = cls.argmax(dim=1)

            correct += (pred == label).sum().item()
            total += label.size(0)

    return correct / total
```

```python
# =========================
# 🔥 IMPORTS
# =========================
import os, cv2, torch, numpy as np
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import torchvision.transforms as transforms
import timm
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import label_binarize
import seaborn as sns
from collections import Counter

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =========================
# 🔥 PATHS
# =========================
train_path = "/content/Acne/Train"
val_path   = "/content/Acne/Validation"

# =========================
# 🔥 PREPROCESSING
# =========================
def enhance_contrast(img):
    img_np = img.permute(1,2,0).cpu().numpy()
    img_np = (img_np * 255).astype(np.uint8)

    lab = cv2.cvtColor(img_np, cv2.COLOR_RGB2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(clipLimit=2.0)
    l = clahe.apply(l)

    enhanced = cv2.merge((l,a,b))
    enhanced = cv2.cvtColor(enhanced, cv2.COLOR_LAB2RGB)

    return torch.tensor(enhanced).permute(2,0,1).float()/255.0


def skin_filter(img):
    img_np = img.permute(1,2,0).cpu().numpy()
    img_np = (img_np * 255).astype(np.uint8)

    hsv = cv2.cvtColor(img_np, cv2.COLOR_RGB2HSV)

    lower = np.array([0, 40, 60])
    upper = np.array([25, 255, 255])

    mask = cv2.inRange(hsv, lower, upper)
    return torch.tensor(mask/255.0).unsqueeze(0).float()


def get_roi(img):
    mask = skin_filter(img)
    return img * mask


# =========================
# 🔥 TRANSFORM
# =========================
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224,224)),
    transforms.ToTensor()
])

# =========================
# 🔥 DATASET (FIXED VERSION)
# =========================
class AcneDataset(Dataset):
    def __init__(self, root_dir):
        self.paths, self.labels = [], []

        for cls in os.listdir(root_dir):
            cls_path = os.path.join(root_dir, cls)

            if not os.path.isdir(cls_path):
                continue

            for img in os.listdir(cls_path):
                img_path = os.path.join(cls_path, img)

                if not os.path.isfile(img_path):
                    continue

                self.paths.append(img_path)

                if cls.lower() == "clear":
                    self.labels.append(0)
                elif cls.lower() == "acne":
                    self.labels.append(1)
                else:
                    self.labels.append(2)

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        img = cv2.imread(self.paths[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        img = transform(img)
        img = enhance_contrast(img)

        label = torch.tensor(self.labels[idx]).long()
        return img, label


# =========================
# 🔥 LOAD DATA
# =========================
train_ds = AcneDataset(train_path)
val_ds   = AcneDataset(val_path)

print("Train:", len(train_ds), " Val:", len(val_ds))

# =========================
# 🔥 CLASS DISTRIBUTION BEFORE
# =========================
labels_names = ["Clear","Acne","Comedo"]
counts_before = Counter(train_ds.labels)

plt.bar(labels_names, counts_before.values())
plt.title("Before Balancing")
plt.show()

# =========================
# 🔥 BALANCING
# =========================
class_counts = np.bincount(train_ds.labels)
weights = 1. / class_counts

sample_weights = [weights[label] for label in train_ds.labels]
sampler = WeightedRandomSampler(sample_weights, len(sample_weights))

train_loader = DataLoader(train_ds, batch_size=16, sampler=sampler)
val_loader   = DataLoader(val_ds, batch_size=16)

# =========================
# 🔥 FEATURE FUSION MODEL
# =========================
class FusionModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.full_net = timm.create_model('mobilenetv3_small_100', pretrained=True, num_classes=0)
        self.roi_net  = timm.create_model('mobilenetv3_small_100', pretrained=True, num_classes=0)

        self.fc = nn.Linear(2048, 256)
        self.out = nn.Linear(256, 3)

    def forward(self, img, roi):
        f1 = self.full_net(img)
        f2 = self.roi_net(roi)

        fused = torch.cat([f1, f2], dim=1)
        fused = self.fc(fused)

        return self.out(fused)


model = FusionModel().to(device)

# =========================
# 🔥 LOSS + OPTIMIZER
# =========================
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=5e-5)

# =========================
# 🔥 TRAIN
# =========================
train_losses, val_accs = [], []

def train_epoch():
    model.train()
    total_loss = 0

    for img, label in train_loader:
        img, label = img.to(device), label.to(device)

        roi = get_roi(img)

        optimizer.zero_grad()
        out = model(img, roi)
        loss = criterion(out, label)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(train_loader)


def evaluate():
    model.eval()
    correct, total = 0, 0

    with torch.no_grad():
        for img, label in val_loader:
            img, label = img.to(device), label.to(device)

            roi = get_roi(img)
            out = model(img, roi)

            pred = out.argmax(dim=1)

            correct += (pred == label).sum().item()
            total += label.size(0)

    return correct / total


# =========================
# 🔥 TRAIN LOOP
# =========================
for epoch in range(5):
    loss = train_epoch()
    acc = evaluate()

    train_losses.append(loss)
    val_accs.append(acc)

    print(f"Epoch {epoch+1}: Loss={loss:.4f}, Acc={acc:.4f}")

# =========================
# 🔥 CONFUSION MATRIX
# =========================
all_preds, all_labels = [], []

model.eval()
with torch.no_grad():
    for img, label in val_loader:
        img = img.to(device)

        roi = get_roi(img)
        out = model(img, roi)

        pred = out.argmax(dim=1).cpu().numpy()

        all_preds.extend(pred)
        all_labels.extend(label.numpy())

cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=labels_names,
            yticklabels=labels_names)
plt.title("Confusion Matrix")
plt.show()

print(classification_report(all_labels, all_preds))

# =========================
# 🔥 LOSS GRAPH
# =========================
plt.plot(train_losses)
plt.title("Training Loss")
plt.show()

# =========================
# 🔥 ACCURACY GRAPH
# =========================
plt.plot(val_accs)
plt.title("Validation Accuracy")
plt.show()

# =========================
# 🔥 ROC CURVE
# =========================
all_probs = []

model.eval()
with torch.no_grad():
    for img, label in val_loader:
        img = img.to(device)

        roi = get_roi(img)
        out = model(img, roi)

        probs = torch.softmax(out, dim=1).cpu().numpy()
        all_probs.extend(probs)

all_probs = np.array(all_probs)
y_true = label_binarize(all_labels, classes=[0,1,2])

plt.figure()

for i in range(3):
    fpr, tpr, _ = roc_curve(y_true[:,i], all_probs[:,i])
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f"{labels_names[i]} (AUC={roc_auc:.2f})")

plt.plot([0,1],[0,1],'--')
plt.legend()
plt.title("ROC Curve")
plt.show()
```
